```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import mean_squared_error
from sklearn.feature_selection import SelectFromModel
```

```python
data = pd.read_csv("OnlineNewsPopularity.csv")
```

```python
data
```

```python
data=data.drop("url", axis =1)
```

```python
data
```

```python
y = data["shares"]
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, y, test_size=0.2)
```

```python
model=LinearRegression()

```

```python
model.fit(data,y)
```

```python
ypre= model.predict(x_train)
ypred=model.predict(x_test)
```

```python
model.score(x_test, ypred)
```

```python
import numpy as np
```

```python
train_residuals = y_train - ypre
summed_train_residuals = np.sum(train_residuals)
print(summed_train_residuals)

test_residuals = y_test - ypred
summed_test_residuals = np.sum(test_residuals)
print(summed_test_residuals)
```

```python
r2_score(y_test,ypred)
```

```python
mean_squared_error(y_test,ypred)
```
